```python
import pandas as pd
```

```python
# Cargar ambos archivos
dataset = pd.read_csv('training_set_features.csv')
dataset_resultados = pd.read_csv('training_set_labels.csv')
```

```python
# Imprimir una parte de ellos
dataset.head()
```

```python
# Imprimir una parte de los resultados
dataset_resultados.head()
```

```python
# Seleccionar sólo una parte de las columnas
columnas = [
    "h1n1_concern", # ¿Qué tanto le preocupa el H1N1 del 0-3?
    "h1n1_knowledge", # ¿Qué tanto sabe del virus del 0-2?
    "behavioral_antiviral_meds", # ¿Cree en antivacunas? 0 o 1
    "behavioral_avoidance", # ¿Evita estar en público? 0 o 1 
    "behavioral_wash_hands", # ¿Lava regularmente sus manos? 0 o 1 
    "behavioral_large_gatherings", # ¿Está en lugares con mucha gente con frecuencia? 0 o 1
    "behavioral_touch_face", # ¿Toca su cara frecuentemente? 0 o 1 
    "doctor_recc_h1n1", # ¿Su doctor recomendó vacunarse? 0 o 1 
    "opinion_h1n1_vacc_effective", # ¿Qué tan efectiva cree que es la vacuna del 1-5?
    "opinion_h1n1_risk", # ¿Qué tanto riesgo cree que tiene la vacuna del 1-5?
    "age_group", # Grupo de edad
    "education", # Máximo grado de estudio
    "sex", # Sexo de la persona
    "income_poverty", # La cantidad de ingresos de la persona
    "employment_status", # Si la persona se encuentra empleada
]

dataset_columnas_interes = dataset[columnas]
```

```python
# Seleccionar sólo las columnas de interés de la vacunación
columnas_resultados = [
    "h1n1_vaccine"
]
dataset_resultados_columnas = dataset_resultados[columnas_resultados]
```

```python
# Obtener el dataset final
df = pd.concat([dataset_columnas_interes, dataset_resultados_columnas], axis=1)
# Borrar filas con datos faltantes
df = df.dropna()
```

```python
# Obtener valores por cada columna
for c in df.columns:
    print(f'{c}: {sorted(df[c].unique().tolist())}')
```

```python
# Remapear valores enteros
texto = ['age_group', 'education', 'sex', 'income_poverty', 'employment_status']

for c in df.columns:
    if c not in texto:
        df[c] = df[c].astype(int)

# Remapear booleanos
booleanos_extra = ['doctor_recc_h1n1', 'h1n1_vaccine']
mapeo_booleanos = { 0: 'f', 1: 't' }
for c in df.columns:
    if c.startswith('behavioral') or c in booleanos_extra:
        df[c] = df[c].map(mapeo_booleanos)

df.head()
```

```python
import itertools
cantidad_filas = 2800
# Crear archivo ARFF
with open('dataset.arff', 'w') as f:
    f.write('@RELATION h1n1')
    
    f.write('\n\n')
    
    # Imprimir lista de atributos
    quote = '"{}"'.format
    for c in df.columns:
        mapf = quote if c in texto else (lambda x: str(x))
        f.write(f"@ATTRIBUTE {c} {{{', '.join(map(mapf, sorted(df[c].unique().tolist())))}}}\n")
    
    f.write('\n')
    f.write('@DATA\n')
    rowf = lambda x: quote(x) if isinstance(x, str) and len(x) > 1 else x
    for r in itertools.islice(df.itertuples(), cantidad_filas):
        f.write(','.join(map(str, map(rowf, r[1:]))))
        f.write('\n')
```
